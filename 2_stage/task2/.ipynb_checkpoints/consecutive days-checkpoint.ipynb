```python
import pandas as pd
import numpy as np

import datetime


df_test = pd.read_csv("data/test.csv")

tvr_18 = "Все 18+_TVR"
tvr_55 = "Все 55+_TVR"

share_18 = "Все 18+_Share"
share_55 = "Все 55+_Share"

df_test["День"] = df_test["День"].apply(lambda x: int(x.split()[1]))
```

```python
dates_test = np.array(df_test["Дата_День"])

df_test["day_datetime"] = df_test["Дата_День"].apply(lambda x: datetime.datetime.strftime(datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"), "%Y-%m-%d"))
df_test["weekday"] = df_test["Дата_День"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").weekday())
```

Подряд идущие дни

```python
consecutive_days = []
for i in range(len(df_test["day_datetime"]) - 1):
    day1, day2 = int(df_test["day_datetime"][i].split('-')[-1]), int(df_test["day_datetime"][i+1].split('-')[-1])
    if day2 - day1 == 1:
        if df_test["day_datetime"][i] not in consecutive_days: consecutive_days.append(df_test["day_datetime"][i]) 
        if df_test["day_datetime"][i+1] not in consecutive_days: consecutive_days.append(df_test["day_datetime"][i+1])        
```
