```python
def get_week(n, days):
    res = []
    i = 0
    while i < n:
        res.append(i // (24 * days))
        i+= 1
    return res
```

```python
import numpy as np
np.array(get_week(100, 7))
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import datetime as DT
```

```python
df_train = pd.read_csv("data/train_result.csv")
df_test = pd.read_csv("data/result/test.csv")
```

```python
df_concat = pd.concat([df_train, df_test])
df_concat["date"] = df_train["Дата"].apply(lambda x: DT.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
df_concat.sort_values(by = ["date"])
```

```python
dates_train = df_train["Дата"].apply(lambda x: DT.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
dates_test = df_test["Дата"].apply(lambda x: DT.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
dates_concat = list(sorted(set(pd.concat([dates_train, dates_test]))))
```

```python
df_weeksmean = pd.DataFrame(list(zip(dates_concat, np.array(get_week(len(dates_concat), 7)))), columns = ["date", "index"])
```

```python
df_weeksmean
```

```python
pd_common = pd.merge(df_concat, df_weeksmean, left_on = "date", right_on = "date")
pd_common
```

```python
tvr_18 = "Все 18+_TVR"
tvr_55 = "Все 55+_TVR"

share_18 = "Все 18+_Share"
share_55 = "Все 55+_Share"

targets = [tvr_18, tvr_55, share_18, share_55]

# print(pd_common[pd_common['index'] == 152])

for target in targets:
    df = pd_common.groupby('index')[target].mean()
#     print(df.loc[152])
    df_weeksmean['mean_index_target_{0}'.format(target)] = df_weeksmean['index'].apply(lambda x: df.loc[x])
```
