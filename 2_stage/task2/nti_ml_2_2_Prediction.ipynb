# Обучение

```python
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import OneHotEncoder

import pandas as pd
import numpy as np

import datetime, math
```

```python
tvr_18 = "Все 18+_TVR"
tvr_55 = "Все 55+_TVR"

share_18 = "Все 18+_Share"
share_55 = "Все 55+_Share"
```

```python
df_train = pd.read_csv("data/train_result.csv")

df_train
```

```python
df_train.columns
```

### Кодирование дней недели

```python
weekdays = ["weekday_{0}".format(i) for i in range(1, 8)]

df_train['weekday'].unique()
```

```python
onehotObject_weekday = OneHotEncoder(handle_unknown = 'ignore')

onehotObject_weekday.fit(np.array(df_train['weekday']).reshape(-1, 1))

onehotObject_weekday.categories_
```

```python
weekday_onehot = pd.DataFrame(onehotObject_weekday.transform(np.array(df_train['weekday']).reshape(-1, 1)).toarray().astype(int), columns = weekdays)
```

```python
weekday_onehot
```

```python
df_train_merge = pd.merge(df_train, weekday_onehot, left_index = True, right_index = True)

df_train_merge.drop("weekday", inplace = True, axis = 1)

df_train_merge
```

```python
df_train = df_train_merge
```

### Кодирование времени

```python
hours_onehot = ["hour_{0}".format(i) for i in range(24)]

df_train["hour"] = df_train["Дата"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").hour)
```

```python
df_train["hour"].value_counts()
```

#### One-hot encoding

```python
onehotObject_time = OneHotEncoder(handle_unknown = 'ignore')
```

```python
onehotObject_time.fit(np.array(df_train["hour"]).reshape(-1, 1))
```

```python
df_train_merge = pd.merge(df_train, pd.DataFrame(onehotObject_time.transform(np.array(df_train["hour"]).reshape(-1, 1)).toarray().astype(int), columns = hours_onehot), left_index = True, right_index = True)
```

```python
df_train_merge
```

```python
df_train = df_train_merge
```

#### Среднее кодирование

```python
hour_mean_tvr18 = df_train.groupby("hour")[tvr_18].mean()
hour_mean_tvr55 = df_train.groupby("hour")[tvr_55].mean()

hour_mean_share18 = df_train.groupby("hour")[share_18].mean()
hour_mean_share55 = df_train.groupby("hour")[share_55].mean()

df_train["hour_mean_tvr18"] = df_train["hour"].apply(lambda x: hour_mean_tvr18[x])
df_train["hour_mean_tvr55"] = df_train["hour"].apply(lambda x: hour_mean_tvr55[x])

df_train["hour_mean_share18"] = df_train["hour"].apply(lambda x: hour_mean_share18[x])
df_train["hour_mean_share55"] = df_train["hour"].apply(lambda x: hour_mean_share55[x])

df_train
```

```python
columns_search = np.array([weekdays, hours_onehot, 'hour_mean_tvr18', 'news_views', 'news_views_shift', 'temp_Moscow',
       'temp_Peter', 'temp_Novosib', 'temp_Ekb', 'searches_wo',
       'searches_tnt', 'searches_1st', 'searches_r1', 'searches_sts',
       'searches_news', 'searches_rentv', 'lockdown_index', 'working_day',
       'oil_price', 'dollar_rate'])
```

```python
errors = []

channel = "ПЕРВЫЙ КАНАЛ"

for column in columns_search:
    print("---{0}---".format(column))
    
    if (type(column) == str): column = [column]
    
    XTrain, XTest, YTrain, YTest = train_test_split(df_train[df_train["Канал"] == channel][column].fillna(0), df_train[df_train["Канал"] == channel][tvr_18], test_size = 0.33, random_state = 42)
    
    model_cbr = CatBoostRegressor(silent = True)
    model_cbr.fit(XTrain, YTrain)
    
    error = mean_squared_log_error(YTest, model_cbr.predict(XTest))
    print("MSLE: {0}".format(error))
    
    errors.append(error)
```

```python
errors_sorted = sorted(errors)

errors_sorted
```

```python
indexes_sorted = list(map(lambda x: errors.index(x), errors_sorted))

indexes_sorted
```

```python
out_indexes = []
current_features = []
all_features = columns_search.copy()

errors = []

models = []

for i in range(len(indexes_sorted)):
    if (type(columns_search[indexes_sorted[i]]) == str):
        current_features.append(columns_search[indexes_sorted[i]])
    else:
        current_features += columns_search[indexes_sorted[i]]

    print(current_features[-1])
        
    XTrain, XTest, YTrain, YTest = train_test_split(df_train[df_train["Канал"] == channel][current_features].fillna(0), df_train[df_train["Канал"] == channel][tvr_18], test_size = 0.33, random_state = 42)
    
    model_cbr = CatBoostRegressor(silent = True)
    model_cbr.fit(XTrain, YTrain)
    
    error = mean_squared_log_error(YTest, np.absolute(model_cbr.predict(XTest)))
    
    print(error)
    
    errors.append(error)
    
    if (i >= 1 and errors[i] > errors[i - 1]):
        print("---making someone out---")
        print("{0} was ejected.".format(current_features[-1]))
        current_features.pop(len(current_features) - 1)
    else:
        models.append(model_cbr)
```

```python
"Last model MSLE error: {0}".format(mean_squared_log_error(YTest, np.absolute(models[-1].predict(XTest))))
```

```python
random_state_ = 42
```

```python
entercolumns_channels = [
    ['searches_news', 'searches_1st', 'working_day'], # Первый канал
    ['searches_news', 'searches_r1', 'working_day'], # Россия 1
    ['searches_wo', 'searches_sts', 'working_day', 'lockdown_index'], # СТС
    ['news_views', 'searches_wo', 'searches_rentv', 'working_day', 'lockdown_index'], # РЕН ТВ
    ['searches_wo', 'searches_tnt', 'working_day', 'searches_sts']] # ТНТ

weekdays = ["weekday_{0}".format(i) for i in range(1, 8)]
hours_onehot = ["hour_{0}".format(i) for i in range(24)]

for channel in entercolumns_channels:
    channel += weekdays
#     channel += hours_onehot
    channel += ['hour_mean_tvr18']
    
channels = df_train["Канал"].unique()

models_tvr18 = []

for channelIndex in range(len(channels)):
    print("---{0}---".format(channels[channelIndex]))
    XTrain, XTest, YTrain, YTest = train_test_split(df_train[entercolumns_channels[channelIndex]][df_train["Канал"] == channels[channelIndex]].fillna(0), df_train[tvr_18][(df_train["Канал"] == channels[channelIndex])], test_size = 0.33, random_state = random_state_)

    model_cbr = CatBoostRegressor(silent = True, random_state = random_state_, loss_function = 'RMSE')
    
#     cat_features_ = hours_onehot + weekdays + ['working_day']
    cat_features_ = weekdays + ['working_day']
    
    model_cbr.fit(XTrain, YTrain, cat_features = cat_features_)
    
    print("MSLE: {0}".format(mean_squared_log_error(YTest, np.absolute(model_cbr.predict(XTest)))))
    print("MSE: {0}".format(mean_squared_error(YTest, model_cbr.predict(XTest))))
    print("R2_SCORE: {0}".format(r2_score(YTest, model_cbr.predict(XTest))))
    models_tvr18.append(model_cbr)
```

```python
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import OneHotEncoder

import pandas as pd
import numpy as np

import datetime, json

from model import Model

tvr_18 = "Все 18+_TVR"
tvr_55 = "Все 55+_TVR"

share_18 = "Все 18+_Share"
share_55 = "Все 55+_Share"

df_train = pd.read_csv("data/train_result.csv")

weekdays = ["weekday_{0}".format(i) for i in range(1, 8)]

onehotObject_weekday = OneHotEncoder(handle_unknown = 'ignore')
onehotObject_weekday.fit(np.array(df_train['weekday']).reshape(-1, 1))
df_train = pd.merge(df_train, pd.DataFrame(onehotObject_weekday.transform(np.array(df_train['weekday']).reshape(-1, 1)).toarray().astype(int), columns = weekdays), left_index = True, right_index = True)
df_train.drop("weekday", inplace = True, axis = 1)


hours_onehot = ["hour_{0}".format(i) for i in range(24)]

df_train["hour"] = df_train["Дата"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").hour)


onehotObject_time = OneHotEncoder(handle_unknown = 'ignore')
onehotObject_time.fit(np.array(df_train["hour"]).reshape(-1, 1))
df_train = pd.merge(df_train, pd.DataFrame(onehotObject_time.transform(np.array(df_train["hour"]).reshape(-1, 1)).toarray().astype(int), columns = hours_onehot), left_index = True, right_index = True)


hour_mean_tvr18 = df_train.groupby("hour")[tvr_18].mean()
hour_mean_tvr55 = df_train.groupby("hour")[tvr_55].mean()

hour_mean_share18 = df_train.groupby("hour")[share_18].mean()
hour_mean_share55 = df_train.groupby("hour")[share_55].mean()

df_train["hour_mean_tvr18"] = df_train["hour"].apply(lambda x: hour_mean_tvr18[x])
df_train["hour_mean_tvr55"] = df_train["hour"].apply(lambda x: hour_mean_tvr55[x])

df_train["hour_mean_share18"] = df_train["hour"].apply(lambda x: hour_mean_share18[x])
df_train["hour_mean_share55"] = df_train["hour"].apply(lambda x: hour_mean_share55[x])


print(df_train)
print("PREPROCESSING COMPLETED")
print("TRAINING")

# best 500
random_state_ = 500

channels = df_train["Канал"].unique()

targets = [tvr_18, tvr_55, share_18, share_55]
```

```python
target = targets[0]

grid = {"n_estimators": [500, 600, 700, 800, 900, 1000, 1100, 1200]}

errors = []

for channelIndex in range(len(channels)):
    print("---{0}---".format(channels[channelIndex]))

    with open("data/models/backup/model_features_{0}_{1}.txt".format(target, channelIndex), "r") as file:
        features = json.loads(file.read())
        
    XTrain, XTest, YTrain, YTest = train_test_split(df_train[df_train["Канал"] == channels[channelIndex]][features].fillna(0), df_train[df_train["Канал"] == channels[channelIndex]][target], test_size = 0.33, random_state = random_state_)

    for grid_i in grid["n_estimators"]:
        # best 800 8
        model_cbr = CatBoostRegressor(random_state = random_state_, silent = True, 
                                      n_estimators = grid_i, 
                                      max_depth = 8,
                                      loss_function = 'RMSE',
                                      eval_metric = 'RMSE', grow_policy = "SymmetricTree")
    #     model_cbr.fit(XTrain, YTrain, plot = True)
        model_cbr.fit(XTrain, YTrain)

    #     model_cbr.grid_search(grid, X = XTrain, y = YTrain, plot = True)

        print("Train MSLE: {0}".format(mean_squared_log_error(YTrain, np.absolute(model_cbr.predict(XTrain)))))
        print("Test MSLE: {0}".format(mean_squared_log_error(YTest, np.absolute(model_cbr.predict(XTest)))))
        errors.append(mean_squared_log_error(YTest, np.absolute(model_cbr.predict(XTest))))
```

```python
import matplotlib.pyplot as plt

grid_len = len(grid["n_estimators"])

for i in range(5):
    fig, ax = plt.subplots()
#     print(errors[5 * i: 5 * (i + 1)])
    ax.plot(grid["n_estimators"], errors[grid_len * i: grid_len * (i + 1)])
```
